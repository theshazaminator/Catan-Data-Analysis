```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\Shawzub\Downloads\archive\scorexharbor.csv")
```

```python
df.head()
```

```python
km = KMeans(n_clusters=1)
y_predicted = km.fit_predict(df[['score','num_of_harbor']])
y_predicted
km.cluster_centers_
```

```python
plt.scatter(df.score, df.num_of_harbor, s=df["count"]*25, c=df.color, alpha=0.5)
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid', s=100)
plt.xlabel("Score")
plt.ylabel('Number of Harbors')
plt.title('Score X Number of Harbors')
plt.legend()
```
